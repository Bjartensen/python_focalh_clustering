```python
import numpy as np
import matplotlib.pyplot as plt

# Generate example data: true and reconstructed energy
np.random.seed(42)
true_energy = np.random.normal(loc=100, scale=10, size=10000)  # True energy values
reconstructed_energy = true_energy + np.random.normal(loc=0, scale=5, size=10000)  # Reconstructed energy with some noise

# Define bins for true and reconstructed energy
bins = np.linspace(70, 130, 31)  # 30 bins between 70 and 130

# Create a 2D histogram
migration_matrix, xedges, yedges = np.histogram2d(
    true_energy, reconstructed_energy, bins=[bins, bins]
)

# Plot the migration matrix
plt.figure(figsize=(8, 6))
plt.imshow(
    migration_matrix.T,  # Transpose to match the orientation of true vs reconstructed
    origin="lower",
    extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],
    aspect="auto",
    cmap="jet",
)
plt.colorbar(label="Number of Events")
plt.xlabel("True Energy")
plt.ylabel("Reconstructed Energy")
plt.title("Migration Matrix: True vs Reconstructed Energy")
plt.plot([70, 130], [70, 130], "r--", label="Perfect Agreement (y = x)")  # Line of perfect agreement
plt.legend()
plt.show()
```
